# Model Building 

```python
df = df.drop(columns='animal name')
```

```python
x = df.iloc[:,:16]
y = df['type']
```

```python
x
```

```python
y
```

```python
#splitting the data
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(xtrain,ytrain)
ypred = knn.predict(xtest)
ypred
```

```python
knn.score(xtrain,ytrain),knn.score(xtest,ytest)
```

Hyperparameter Tuning 

```python
trainac = []
testac = []

for i in range(1,25):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    trainac.append(knn.score(xtrain,ytrain))
    testac.append(knn.score(xtest,ytest))
```

```python
trainac
```

```python
testac
```

```python
#plot
plt.plot(range(1,25), trainac,linestyle = '-.', color = 'blue',label = 'Training Accuracy')
plt.plot(range(1,25),testac, linestyle = '--', color = 'green', label = 'Testing Accuracy')
plt.xlabel('Number of K')
plt.ylabel('Accuracy')
plt.legend()
```

Final Model

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(xtrain,ytrain)
knn.score(xtrain,ytrain),knn.score(xtest,ytest)
```
